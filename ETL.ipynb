### Archivo 4: Accesos a internet fijo por tecnología y localidad

```python
df4 = pd.read_csv('./datastreams/accesos_a_internet_fijo_por_tecnología_y_localidad.csv', decimal=',', thousands='.')
```

```python
df4.head()
```

```python
df4.info()
```

```python
df4.drop(columns='Link Indec', inplace=True)
```

```python
df4.to_csv('./datasets/accesos_a_internet_fijo_por_tecnologia_y_localidad.csv')
```

En este archivo reemplazamos valores -0 por 0, reemplazamos las ',' por '.' para miles los '.' por vacio. Eliminamos las columnas Unnamed: 13 y Link Indec porque consideramos que no son de utilidad para el analisis. 

### Archivo 5: Accesos a internet fijo por velocidad bajada y provincia

```python
df5 = pd.read_csv('./datastreams/accesos_a_internet_fijo_por_velocidad_bajada__y_provincia.csv', decimal=',', thousands='.')
```

```python
df5.head()
```

```python
df5.info()
```

```python
df5['Año'].unique()
```

```python
df5[df5['Año'].isna()]
```

```python
df5 = df5.dropna(subset=['Año', 'Trimestre', 'Provincia'], how= 'all')
```

```python
df5['Año'].unique()
```

```python
reemplazo_anios = {20220.: 2022, 20210.: 2021, 20200.: 2020, 20190.: 2019, 20180.: 2018, 20170.: 2017}
```

```python
df5['Año'] = df5['Año'].replace(reemplazo_anios).astype(int)
```

```python
df5['Trimestre'].unique()
```

```python
reemplazo_trimestres = {40.: 4, 30.: 3, 20.: 2, 10.: 1}
```

```python
df5['Trimestre'] = df5['Trimestre'].replace(reemplazo_trimestres).astype(int)
```

```python
df5 = df5.apply(lambda x: x.map({ '- 0': 0 }).fillna(x))
```

```python
df5 = df5.replace(r'\.', '', regex=True)
```

```python
columnas_numericas = df5.columns.difference(['Provincia'])
```

```python
df5[columnas_numericas] = df5[columnas_numericas].apply(pd.to_numeric, errors='coerce', downcast='integer')
```

```python
df5.info()
```

```python
df5.to_csv('./datasets/accesos_a_internet_fijo_por_velocidad_bajada_y_provincia.csv')
```

Se suiguio las mismas metodologia que en los archivos anteirores para realizar la limpieza de datos
