```python
import pandas as pd
```

```python
df1 = pd.read_csv('../data_sql/Acceso_a_banda_ancha_y_banda_angosto_por_provincia.csv', decimal=',', thousands='.')
```

```python
df1['provincia_id'] = pd.factorize(df1['Provincia'])[0]
df1['año_id'] = pd.factorize(df1['Año'])[0]
```

```python
df1.head()
```

```python
df1.drop(columns='Unnamed: 0', inplace=True)
```

```python
df1['Dial up'].fillna(0, inplace=True)
```

```python
df1.info()
```

```python
df1.to_csv('../data_sql/sql_acc_bandas_provincias.csv', index=False)
```
